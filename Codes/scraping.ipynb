### <center> Récupération des liens de tous les pokemons

```python

import urllib.request 
import time


class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/71.0"
opener= AppURLopener()

response = opener.open("https://pokemondb.net/pokedex/national")
soup = BeautifulSoup(response, 'html.parser')
```

```python
liste_liens  =[]
attributs = soup.select("a[class='ent-name']")
for link in attributs:
    lien=(link.get("href"))
    lien = "https://pokemondb.net" + lien
    liste_liens.append(lien)
liste_liens


Liens = pd.DataFrame.from_dict( {"Liens" : liste_liens})
Liens
```

```python
test = Liens.iloc[0:12]
```

### <center>Récuparation du tableau Pokedex Data

```python
Nom = []
national_n = []
types = []
species = []
height = []
weight = []
abilities = []
local_number = []
liste_url = []




for i in range(len(test)):
    liste_url.append(test.iloc[i]["Liens"])
    class AppURLopener(urllib.request.FancyURLopener):
        version = "Mozilla/71.0"
    opener= AppURLopener()
    
    response = opener.open(urllib.parse.unquote(liste_url[i]))
    soup = BeautifulSoup(response, "html.parser")
    #################################### Nom ################################################
    
    nom  = soup.find("h1").getText()
    Nom.append(nom) 
    #################################### National N° ################################################

    attribut_training_nn = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[0]
    str_cells_nn = str(attribut_training_nn)
    clean_nn = BeautifulSoup(str_cells_nn, "lxml").get_text().strip() 
    national_n.append(clean_nn)
    #################################### Type ################################################
    attribut_training_type = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[1]
    str_cells_type = str(attribut_training_type)
    clean_type = BeautifulSoup(str_cells_type, "lxml").get_text().strip() 
    types.append(clean_type)
    #################################### Species ################################################
    attribut_training_sp = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[2]
    str_cells_sp = str(attribut_training_sp)
    clean_sp = BeautifulSoup(str_cells_sp, "lxml").get_text().strip() 
    species.append(clean_sp)
    
    #################################### Height ################################################
    attribut_training_he = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[3]
    str_cells_he = str(attribut_training_he)
    clean_he = BeautifulSoup(str_cells_he, "lxml").get_text().strip() 
    height.append(clean_he)
    #################################### Weight ################################################
    attribut_training_we = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[4]
    str_cells_we = str(attribut_training_we)
    clean_we = BeautifulSoup(str_cells_we, "lxml").get_text().strip() 
    weight.append(clean_we)
    #################################### Abilities ################################################
    attribut_training_ab = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[5]
    str_cells_ab = str(attribut_training_ab)
    clean_ab = BeautifulSoup(str_cells_ab, "lxml").get_text().strip() 
    abilities.append(clean_ab)
    #################################### Local N° ################################################
    attribut_training_lo = soup.findAll('table', {'class' : 'vitals-table'})[0].findAll('td')[6]
    str_cells_lo = str(attribut_training_lo)
    clean_lo = BeautifulSoup(str_cells_lo, "lxml").get_text().strip() 
    local_number.append(clean_lo)
    
    
    
    
    
    
    
    
tab=np.array([Nom,national_n,types, species, height, weight, abilities, local_number])
data=pd.DataFrame(tab, index= ['Nom', 'National N°','Type','Species', 'Height', 'Weight', 'Abilities','Local N°'])
dft = pd.DataFrame.transpose(data)
dft

    
```

### <center>Récupération du  tableau Training

```python
Nom = []
ev_yield = []
catch_rate = []
base_friendship = []
base_exp = []
growth_rate = []
liste_url = []





for i in range(len(test)):
    liste_url.append(test.iloc[i]["Liens"])
    class AppURLopener(urllib.request.FancyURLopener):
        version = "Mozilla/71.0"
    opener= AppURLopener()
    
    response = opener.open(urllib.parse.unquote(liste_url[i]))
    soup = BeautifulSoup(response, "html.parser")
    #################################### Nom ################################################
    
    nom  = soup.find("h1").getText()
    Nom.append(nom) 
    #################################### EV Yield ################################################

    attribut_training_ev = soup.findAll('table', {'class' : 'vitals-table'})[1].findAll('td')[0]
    str_cells_ev = str(attribut_training_ev)
    clean_ev = BeautifulSoup(str_cells_ev, "lxml").get_text().strip() 
    ev_yield.append(clean_ev)
    #################################### Catch rate ################################################
    attribut_training_cr = soup.findAll('table', {'class' : 'vitals-table'})[1].findAll('td')[1]
    str_cells_cr = str(attribut_training_cr)
    clean_cr = BeautifulSoup(str_cells_cr, "lxml").get_text().strip() 
    catch_rate.append(clean_cr)
    #################################### Base Friendship ################################################
    attribut_training_bf = soup.findAll('table', {'class' : 'vitals-table'})[1].findAll('td')[2]
    str_cells_bf = str(attribut_training_bf)
    clean_bf = BeautifulSoup(str_cells_bf, "lxml").get_text().strip() 
    base_friendship.append(clean_bf)
    
    #################################### Base Exp ################################################
    attribut_training_be = soup.findAll('table', {'class' : 'vitals-table'})[1].findAll('td')[3]
    str_cells_be = str(attribut_training_be)
    clean_be = BeautifulSoup(str_cells_be, "lxml").get_text().strip() 
    base_exp.append(clean_be)
    #################################### Growth rate ################################################
    attribut_training_gr = soup.findAll('table', {'class' : 'vitals-table'})[1].findAll('td')[4]
    str_cells_gr = str(attribut_training_gr)
    clean_gr = BeautifulSoup(str_cells_gr, "lxml").get_text().strip() 
    growth_rate.append(clean_gr)
    
    
    
    
    
    
    
    
tab=np.array([Nom,ev_yield,catch_rate, base_friendship, base_exp, growth_rate])
data=pd.DataFrame(tab, index= ['Nom', 'Ev Yield','Catch Rate','Base Friendship', 'Base Exp', 'Growth Rate'])
dft = pd.DataFrame.transpose(data)
dft

    
    
```

### <center> Récupération des images des pokemons

```python
liste_url = []
liste_image = []
Nom = []
import requests

for i in range(len(test)):
    liste_url.append(test.iloc[i]["Liens"])
    class AppURLopener(urllib.request.FancyURLopener):
        version = "Mozilla/71.0"
    opener= AppURLopener()
    
    response = opener.open(urllib.parse.unquote(liste_url[i]))
    soup = BeautifulSoup(response, "html.parser")
    nom  = soup.find("h1").getText()
    Nom.append(nom) 
    
    image = soup.findAll('img')[0]['src']
    liste_image.append(image)
    
    for i in range(len(liste_image)):
        with open("Images/"+Nom[i]+".jpg", 'wb') as handle:
                response = requests.get(liste_image[i], stream=True)

                if not response.ok:
                    print(response)

                for block in response.iter_content(1024):
                    if not block:
                        break

                    handle.write(block)
```
